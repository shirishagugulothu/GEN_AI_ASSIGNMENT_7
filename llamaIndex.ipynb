```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.huggingface import HuggingFaceLLM
# SimpleDirectoryReader is used to read documents and VectorStoreIndex is used for indexing the documents
# into vectors and settings is used to specify that the system should use the specify llm model and embedding modal
#HuggingFaceEmbedding is used to specify the embedding model and huggingFaceLLm is used to specify the LLm model
```

```python
#in this cell we are using the file operator in the wich we will be opening the document that is in write
#format and in f.write we will be writting the content

with open('demo.txt', 'w') as f:
    demo = f.write("""Human activity since the Industrial Revolution has dramatically raised concentrations of greenhouse gases—especially carbon dioxide, methane,
    and nitrous oxide—in the atmosphere. This enhancement of the greenhouse effect traps more heat, causing Earth’s average temperature to rise.
    The primary sources include burning fossil fuels for energy and transport, deforestation that reduces CO₂ absorption, and agriculture that produces methane and nitrous oxide.
    Industrial processes like cement and steel production further contribute significant CO₂ emissions. Although natural factors like volcanic activity, solar variations,
    and Earth’s orbital changes affect climate over millennia, they cannot explain the rapid warming seen over the past century""")

#opens the docs and analyses the info in it by nothing it down
```

```python
document_store=SimpleDirectoryReader(".").load_data() #reads the input doc
```

```python
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")
#here we will assign the embedding model
```

```python
llm=HuggingFaceLLM(
    model_name="gpt2",
    tokenizer_name="gpt2"
)
```

```python
Settings.embed_model=embed_model
Settings.llm=llm  #sets the process/machanism globally
```

```python
index=VectorStoreIndex(document_store) # complete indexing processing is done in this step
```

```python
qa=index.as_query_engine() #updates or changes index as a query engine (this lets us to ask question and generate answer)
```

```python
import re # to clean the text, removing special chars and compressing the output (regular expression)
def clean_output(response):
  text=str(response).strip()
  first_sentence=re.split(r"\n",text)[0].strip()
  return first_sentence

print("Answer :",clean_output(qa.query("How many greenhouse gases are there?")))
#.quer(...) is a method of the query engine that:
#take your input question.
#Converts it into an embedding (via embed_model)
#Searches the index for similar document chunks.
#Sends thee most relevant context + your  question to the LLM(via LLM).
#Return a response generated by the LLM.
```
